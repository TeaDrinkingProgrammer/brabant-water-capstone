```python
import pandas as pd

df = pd.read_parquet('../data/rucphen_precipitation.parquet')
df.rename(columns={'polygon_1663': 'precipitation'}, inplace=True)
df.set_index('timestamp', inplace=True)
df.fillna(0, inplace=True)
df['precipitation'] = df['precipitation'].apply(lambda x: x if x >= 0 else 0)
df.head()
```

```python
df['precipitation'].plot()
```

```python
df.rolling(4).sum()['precipitation'].plot()
```
